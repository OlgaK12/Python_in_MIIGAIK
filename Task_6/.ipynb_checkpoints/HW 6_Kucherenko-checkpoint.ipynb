```python
time=df.groupby(df.crash_time).size().nlargest(1).reset_index(name='count').crash_time[0]
#time='23:39:07'
def a(time):
    if type(time)==str:
        if len(time)==8:
            time=int(time[0:2])*3600+int(time[3:5])*60+int(time[6:8])
        else:
            time=int(time[0:1])*3600+int(time[2:4])*60+int(time[5:7])
    else:
        h=time//3600 if time//3600>9 else '0'+str(time//3600)
        m=time%3600//60 if time%3600//60>9 else '0'+str(time%3600//60)
        s=time%3600%60 if time%3600%60>9 else '0'+str(time%3600%60)
        time=str(h)+':'+str(m)+':'+str(s)     
    return time
a(time)
```

```python
times=df.groupby(df.crash_time).size().reset_index(name='count')
times['crash_time']=times['crash_time'].apply(lambda x: a(x))
times.sort_values('crash_time')['count'].plot.bar()
```
