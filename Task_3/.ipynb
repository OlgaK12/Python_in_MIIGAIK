### Задача 1 Сортировка (1 балл)
На вход поступает список из вещественных чисел. Отсортировать его по возрастанию, элементы в отсортированном списке должны иметь 3 знака после запятой.
<br> **Входные данные**: [1.0,-2.0,3.0].<br> **Выходные данные**: [-2.000,1.000,3.000]

```python
#решение
a =input()
a=a.split()
a.sort()
for i in range(len(a)):
    a[i]=("%.3f" % float(a[i]))
print(a)
```

### Задача 2 Средний символ (1 балл)
На вход поступает строка. Ваша задача - вернуть средний символ строки. Если длина строки нечетна, верните средний символ. Если длина строки четная, верните средние 2 символа.
<br> **Входные данные** : "test". **Выходные данные**: "es"
<br> **Входные данные** : "sunsets". **Выходные данные**: "s"

```python
#решение
a=input()
if (len(a) % 2 == 0): #если длина строки четная
    print(a[int(len(a)/2-1)], a[int(len(a)/2)], sep='')
else: #иначе (если длина строки нечетная)
    print(a[int(len(a)/2)])
```

### Задача 4 ДНК (1 балл)
На вход подается строка "ДНК". Ваша цель построить комплиментарную цепь к этой ДНК. Материал для [справки](https://bit.ly/2gyHYCV), как меняются отдельные азотистые основания.

<br> **Входные данные**: "GTAT". **Выходные данные**:"CATA"

<br>**Входные данные**:list = ["cat","city","exception","dog","exception","kick"], $k = 2$. **Выходные данные**: "cityexception"

```python
#решение
list=(input('введите список слов через пробел: '))
k=input('введите количество слов: ')
print(int(k))
new_list=[]
list=list.split()
if int(k)<0 or len(list)==0 or int(k)>len(list):
    print('')
else:
    for i in range(len(list)-int(k)+1):
        new_list.append(len(list[i]))
        #print(new_list) # для наглядности, как работает цикл
        for j in range (1, int(k)):
            new_list[i]=new_list[i]+len(list[i+j])
            #print(new_list) # для наглядности, как работает цикл
            
    #print(new_list) # для наглядности
    #print(max(new_list)) # для наглядности
    #print(new_list.index(max(new_list))) # для наглядности
    print("".join(list[new_list.index(max(new_list)):new_list.index(max(new_list))+int(k)])) # при просмотре списка [от:до] список не показывает последний член "до". То есть он показывает факктически [от:до-1]
#print(list)
```

### Задача 8 Смайлики (1 балла)
На вход подается список из смайликов. Правильные смайлики имеют следующую структуру: глаза это ":" или ";", нос - "", "-" или "~", улыбка - ")" или "D". Ваша цель посчитать количество таких смайликов в списке. Пример правильных смайликов: ":-D", ":)", ";-D" ,":~D". 
    <br> **Входные данные**: [':)', ';(', ';}', ':-D']. **Выходные данные**: 2

```python
#решение
list=input('Введине список смайликов через пробел: ')
list=list.split()
print(list)
count_smile=(0)
for i in range(len(list)):
    if(list[i][0]==':' or list[i][0]==';'):
        if(list[i][1]=='-' or list[i][1]=='~'):
            if(list[i][2]==')' or list[i][2]=='D'):
                count_smile=count_smile+1
        if(list[i][1]==')' or list[i][1]=='D'):
                count_smile=count_smile+1
print(count_smile)
```
